```python
import pandas as pd
import random
```

```python
aut_2024 = pd.read_csv('datasets/daep/Austria/Austria_ENTSO-E_2024.csv')
# Check data types
print("Data types:")
print(aut_2024.dtypes)

# Check timestamp format
print("\nTimestamp sample:")
print(aut_2024['Timestamp'].head())

# Check for missing values
print("\nMissing values:")
print(aut_2024.isna().sum())

# Check for basic statistics of the Austria column
print("\nAustria column statistics:")
print(aut_2024['Austria'].describe())
```

```python
print(aut_2024.dtypes)
```

```python
# Convert Timestamp column to datetime
aut_2024['Timestamp'] = pd.to_datetime(aut_2024['Timestamp'])

# Check that conversion worked
print("Updated data types:")
print(aut_2024.dtypes)

display(aut_2024.head(20))
```
